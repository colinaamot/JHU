```python
import matplotlib.pyplot as plt

# Data for US population by decade
decades = ['1950', '1960', '1970', '1980', '1990', '2000', '2010', '2020']
population = [151325798, 179323175, 203211926, 226545805, 248709873, 281421906, 308745538, 331002651]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(decades, population, marker='o', linestyle='-', color='b')

# Customize the plot
plt.title('US Population by Decade')
plt.xlabel('Decade')
plt.ylabel('Population')
plt.grid(True)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for COVID-19 infections by month in 2020
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
infections = [0, 11, 24304, 884663, 1813547, 1905029, 4626383, 5483946, 4594944, 8938436, 4445105, 6756805]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(months, infections, marker='o', linestyle='-', color='b')

# Customize the plot
plt.title('COVID-19 Infections by Month in 2020')
plt.xlabel('Month')
plt.ylabel('Infections')
plt.grid(True)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute hyperbolic cosine of values
x = np.linspace(0, 10, 100)
y = np.cosh(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='b')

# Customize the plot
plt.title('Hyperbolic Cosine')
plt.xlabel('x')
plt.ylabel('cosh(x)')
plt.grid(True)

# Display the plot
plt.show()
```
